# **1) Create a Laptop class with a private attribute __price. Implement a method set_price(new_price) to update the price and get_price() to retrieve the price.**

```python
class Laptop:
    def __init__(self, price):
        self.__price = price  # Private attribute

    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price

    def get_price(self):
        return self.__price

laptop = Laptop(1000)
print("Initial Price =", laptop.get_price())
laptop.set_price(1200)
print("Updated Price =", laptop.get_price())
```

# **2) Create a BankAccount class with a private attribute __balance. Implement methods to deposit, withdraw, and check balance while ensuring the balance cannot be modified directly**

```python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

# Example usage
account = BankAccount(500)
print("Total Balance:", account.get_balance())
account.deposit(200)
print("deposit:", account.get_balance())
account.withdraw(100)
print("withdrawal:", account.get_balance())
```

# **4) Create a Book class with private attributes __title and __author. Implement a constructor to set these values and provide methods to retrieve them.**

```python
class Book:
    def __init__(self, title, author):
        self.__title = title  # Private attribute
        self.__author = author  # Private attribute

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

book = Book("1984", "George Orwell")
print("Title = ", book.get_title())
print("Author = ", book.get_author())
```

# **5) Create a Car class with a private attribute __speed. Implement methods to increase speed, decrease speed, and get the current speed, ensuring speed does not go below 0.**

```python
class Car:
    def __init__(self, speed=0):
      self.__speed = max(speed, 0)
    def change_speed(self, amount):
      self.__speed = max(0, self.__speed + amount)
    def get_speed(self):
      return self.__speed



car = Car()
print(car.get_speed())

car.change_speed(40)
print(car.get_speed())

car.change_speed(20)
print(car.get_speed())

car.change_speed(-40)
print(car.get_speed())
```
